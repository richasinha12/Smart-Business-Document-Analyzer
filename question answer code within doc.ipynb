```python
import os
import uuid
import base64
import gc
import torch
from IPython import display
from unstructured.partition.pdf import partition_pdf
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.schema.document import Document
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Mount Google Drive for saving/loading models if needed
from google.colab import drive
drive.mount('/content/drive')

output_path = "./images"

# Get elements
raw_pdf_elements = partition_pdf(
    filename="/content/anti canal oder.pdf",
    extract_images_in_pdf=True,
    infer_table_structure=True,
    chunking_strategy="by_title",
    max_characters=4000,
    new_after_n_chars=3800,
    combine_text_under_n_chars=2000,
    extract_image_block_output_dir=output_path,
)

# Initialize Hugging Face pipeline for summarization
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(text_elements, summaries):
    for e in text_elements:
        summary = summarizer(e, max_length=150, min_length=30, do_sample=False)
        summaries.append(summary[0]['summary_text'])

def process_images(output_path, image_elements, image_summaries):
    image_model = "Salesforce/blip-image-captioning-base"
    image_pipeline = pipeline("image-to-text", model=image_model)

    def encode_image(image_path):
        with open(image_path, "rb") as f:
            return base64.b64encode(f.read()).decode('utf-8')

    def summarize_image(encoded_image):
        results = image_pipeline(encoded_image)
        return results[0]['generated_text'] if results else "No description available."

    for i in os.listdir(output_path):
        if i.endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(output_path, i)
            encoded_image = encode_image(image_path)
            image_elements.append(encoded_image)
            summary = summarize_image(encoded_image)
            image_summaries.append(summary)

# Separate lists for each type of document
text_elements = []
table_elements = []
text_summaries = []
table_summaries = []

# Extract text and table elements
for e in raw_pdf_elements:
    if 'CompositeElement' in repr(e):
        text_elements.append(e.text)
    elif 'Table' in repr(e):
        table_elements.append(e.text)

# Summarize text and table elements
summarize_text(text_elements, text_summaries)
summarize_text(table_elements, table_summaries)

# Process images
image_elements = []
image_summaries = []
process_images(output_path, image_elements, image_summaries)

# Create Documents and Vectorstore
def create_documents_and_vectorstore(text_elements, text_summaries, table_elements, table_summaries, image_elements, image_summaries):
    documents = []
    retrieve_contents = []

    def add_documents(elements, summaries, doc_type):
        for e, s in zip(elements, summaries):
            doc_id = str(uuid.uuid4())
            doc = Document(
                page_content=s,
                metadata={
                    'id': doc_id,
                    'type': doc_type,
                    'original_content': e
                }
            )
            retrieve_contents.append((doc_id, e))
            documents.append(doc)

    add_documents(text_elements, text_summaries, 'text')
    add_documents(table_elements, table_summaries, 'table')
    add_documents(image_elements, image_summaries, 'image')

    vectorstore = FAISS.from_documents(documents=documents, embedding=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2"))
    vectorstore.save_local("/content/drive/MyDrive/faiss_index")  # Save to Google Drive

    return vectorstore

vectorstore = create_documents_and_vectorstore(text_elements, text_summaries, table_elements, table_summaries, image_elements, image_summaries)

# Initialize LLM for question answering
prompt_template = """
You are a vet doctor and an expert in analyzing dog's health.
Answer the question based only on the following context, which can include text, images, and tables:
{context}
Question: {question}
Don't answer if you are not sure and decline to answer and say "Sorry, I don't have much information about it."
Just return the helpful answer in as much detail as possible.
Answer:
"""

model_name = "google/flan-t5-base"  # Use a smaller model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

pipe = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=1024
)

llm = HuggingFacePipeline(pipeline=pipe)
qa_chain = LLMChain(llm=llm, prompt=PromptTemplate.from_template(prompt_template))

def answer(question):
    relevant_docs = vectorstore.similarity_search(question)
    context = ""
    relevant_images = []

    for d in relevant_docs:
        if d.metadata['type'] == 'text':
            context += '[text]' + d.metadata['original_content']
        elif d.metadata['type'] == 'table':
            context += '[table]' + d.metadata['original_content']
        elif d.metadata['type'] == 'image':
            context += '[image]' + d.page_content
            relevant_images.append(d.metadata['original_content'])

    result = qa_chain.run({'context': context, 'question': question})
    return result, relevant_images

# Sample question for testing
result, relevant_images = answer("What is Gingivitis?")
print(result)

# Display the first image if available
if relevant_images:
    display.display(display.Image(base64.b64decode(relevant_images[0])))

# Clear GPU memory if necessary
gc.collect()
torch.cuda.empty_cache()
```
